```python
from sec_api import QueryApi
import time

# Initialize with your API key
queryApi = QueryApi("45f3eb0fe75cfb2a09850b9658032aecf08a9e2b9ed4e60fae0456043a75cefa")

tickers = [
    "AMZN", "AXP", "AMGN", "AAPL", "BA", "CAT", "CSCO", "CVX", 
    "GS", "HD", "HON", "IBM", "JNJ", "KO", "JPM", "MCD", "MMM", 
    "MRK", "MSFT", "NKE", "PG", "SHW", "TRV", "UNH", "CRM", "NVDA", 
    "VZ", "V", "WMT", "DIS"
]
ticker_str = ",".join(tickers)
results = []
for i in range(10):
    ticker = tickers[3*i:3*i+3]
    ticker_str = ",".join(ticker)
    try:
        # Execute query
        result = queryApi.get_filings({
            "query": {
                "query_string": {
                    "query": f"formType:\"10-K\" AND ticker:({ticker_str})"
                }
            },
            "from": "0",
            "size": "50",
            "sort": [{"filedAt": {"order": "desc"}}]
        })
        results.append(result)
        if result['filings']:
            for filing in result['filings']:
                print(filing['linkToFilingDetails'])

        
    except Exception as e:
        print(f"Error retrieving 10-K for {ticker}: {str(e)}")
        
```

```python
import pathlib
import codecs
output_dir = pathlib.Path('api-data')
output_dir.mkdir(exist_ok=True)
with codecs.open(f'{output_dir}/10-Ks.txt', 'w', 'utf-8', errors='ignore') as f:
    for result in results:
        for filing in result['filings']:
            f.write(filing + '\n')
```

```python
import json

def generate_file(tick_dict):
    year = int(tick_dict['filedAt'][:4])-1
    output_dir = pathlib.Path(f'api-data/{year}/{tick_dict["ticker"]}')
    output_dir.mkdir(parents=True, exist_ok=True)
    with codecs.open(f'{output_dir}/10_K_info.txt', 'w', 'utf-8', errors='ignore') as f:
        json.dump(tick_dict, f)

def read_file(year, ticker):
    output_dir = pathlib.Path(f'api-data/{year}/{ticker}')
    with codecs.open(f'{output_dir}/10_K_info.txt', 'r', 'utf-8', errors='ignore') as f:
        return json.load(f)
    
```
